### TRAIN AND TEST SPLIT 

```python
x= df2.drop('price' , axis =1)
y = df2['price']
```

```python
corrmatrix=df2.corr()
#corrmatrix
plt.subplots(figsize=(20,20))
sns.heatmap(corrmatrix,vmin=-1,vmax=1,annot=True,linewidth=.2,cmap='YlGnBu')
```

```python

import numpy as np

# correlation matrix
corr = df2.corr().abs()

# correlation with target
price_corr = corr['price']

to_drop = set()

for col in corr.columns:
    for row in corr.index:
        if col != row and corr.loc[row, col] >= 0.8:
            # drop the one less correlated with price
            if price_corr[row] < price_corr[col]:
                to_drop.add(row)
            else:
                to_drop.add(col)

# final reduced dataframe
df2 = df2.drop(columns=to_drop)

print("Dropped columns:", to_drop)
print("Remaining columns:", df2.shape[1])
```

```python

from sklearn.preprocessing import StandardScaler

normalisation = StandardScaler()
x=normalisation.fit_transform(x)
# Coverting to Dataframe
x=pd.DataFrame(x)
x.describe()
```

```python

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,random_state =42,test_size=0.33)
```

## MODELS

### LINEAR REGRESSION 

```python

from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
lr_model.fit(x_train, y_train)
y_pred1 = lr_model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(y_test, y_pred1)
mse = mean_squared_error(y_test, y_pred1)
r2 = r2_score(y_test, y_pred1)
print(f'THE SCORE ARE MAE {mae} , MSE {mse} , R2 is {r2} ')
```

### DECISION REGRESSOR 

```python

from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(x_train,y_train)
y_pred2 = dtr.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(y_test, y_pred2)
mse = mean_squared_error(y_test, y_pred2)
r2 = r2_score(y_test, y_pred2)
print(f'THE SCORE ARE MAE {mae} , MSE {mse} , R2 is {r2} ')
```

### RANDOM FOREST

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor( n_estimators=600,max_depth=40, min_samples_split=2,min_samples_leaf=1,max_features='sqrt',random_state=42,n_jobs=-1 )
rfr.fit(x_train,y_train)
y_pred3 = rfr.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(y_test, y_pred3)
mse = mean_squared_error(y_test, y_pred3)
r2 = r2_score(y_test, y_pred3)
print(f'THE SCORE ARE MAE {mae} , MSE {mse} , R2 is {r2} ')
```

### GRADIENT BOOSTING REGRESSOR 

```python

from sklearn.ensemble import GradientBoostingRegressor
gb_reg = GradientBoostingRegressor()
gb_reg.fit(x_train, y_train)
y_pred4=gb_reg.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(y_test, y_pred4)
mse = mean_squared_error(y_test, y_pred4)
r2 = r2_score(y_test, y_pred4)
print(f'THE SCORE ARE MAE {mae} , MSE {mse} , R2 is {r2} ')
```

### SUPPORT VECTOR REGRESSOR

```python
from sklearn.svm import SVR
svm = svr = SVR(kernel='rbf', C=100, epsilon=0.1, gamma='scale')
svm.fit(x_train , y_train)
y_pred5 = svm.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(y_test, y_pred5)
mse = mean_squared_error(y_test, y_pred5)
r2 = r2_score(y_test, y_pred5)
print(f'THE SCORE ARE MAE {mae} , MSE {mse} , R2 is {r2} ')
```
